```python
# Part one

import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')

def func_y(x):
    y = x**2 - 4*x + 2
    return y

def gradient_descent(previous_x, learning_rate, epoch):
    x_gd = []
    y_gd = []
    
    x_gd.append(previous_x)
    y_gd.append(func_y(previous_x))

    # begin the loops to update x and y
    for i in range(epoch):
        current_x = previous_x - learning_rate*(2*previous_x - 4)
        x_gd.append(current_x)
        y_gd.append(func_y(current_x))

        # update previous_x
        previous_x = current_x

    return x_gd, y_gd

# Initialize x0 and learning rate
x0 = 4
learning_rate = 0.15
epoch = 10 # Number of tries

# Set the function
x = np.arange(-1, 5, 0.01)
y = func_y(x)
```

```python
# Plot the function 
def plot_fun(x, y):
    plt.figure(figsize=(20, 10))
    plt.plot(x, y, lw = 0.9, color = 'k')
    plt.xlim([min(x), max(x)])
    plt.ylim([-3, max(y)+1])
    plt.ylabel('y')
    plt.xlabel('X')
    plt.title('y = x**2 - 4*x + 2')

plot_fun(x, y);
```

```python
def plot_dec_fun(x, y, alpha=learning_rate):
    plt.figure(figsize=(20, 10))
    plt.plot(x, y, lw = 0.9, color = 'k', label='Y')
    plt.xlim([min(x), max(x)])
    plt.ylim([-3, max(y)+1])
    plt.title('Gradient Decent for Theta')
    plt.ylabel('F(Theta)')
    plt.xlabel('Theta')
    x_gd, y_gd = gradient_descent(x0, alpha, epoch)
    plt.scatter(x_gd, y_gd, c = 'b', label='Gradient Descent')
    plt.legend(loc='best')

# Try different learning rates    

plot_dec_fun(x, y, alpha=.001)
# plot_dec_fun(x, y, alpha=.01)
# plot_dec_fun(x, y, alpha=.1)
```

```python
from mpl_toolkits.mplot3d import axes3d, Axes3D 

# Part 2
def f(x, y):
    return 4*x**2 + 2*y**2 - 2*x*y

x = np.linspace(-50, 50, 50)
y = np.linspace(-50, 50, 50)

def plot_3d(x, y):
    X, Y = np.meshgrid(x, y)
    Z = f(X, Y)
    fig = plt.figure(figsize=(20, 10))
    ax = plt.axes(projection='3d', alpha=.3)
    ax.scatter3D(X, Y, Z, depthshade=True)
    ax.set_title('z = 4*x**2 + 2*y**2 - 2*x*y')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z');

plot_3d(x, y)
```

```python
# Our cost functions for x and y
# First partial derivative with respect to x and y
def dx (x, y):
    return 8*x - 2*y
def dy (x, y):
    return 4*y - 2*x 

def gradient_descent_2():

    # Create gradient arrays
    grad_x = [] 
    grad_y = []
    grad_z = []

    # Our initinal guess
    theta_0  = 25
    theta_1  = 35

    alpha = .05
    epoch = 10000

    grad_x.append(theta_0)
    grad_y.append(theta_1)
    grad_z.append(f(theta_0, theta_1))

    # Run the gradient
    for i in range(epoch):
        current_theta_0 = theta_0 - alpha * dx(theta_0, theta_1)
        current_theta_1 = theta_1 - alpha * dy(theta_0, theta_1)
        grad_x.append(current_theta_0)
        grad_y.append(current_theta_1)
        grad_z.append(f(current_theta_0, current_theta_1))

        # Update
        theta_0 = current_theta_0
        theta_1 = current_theta_1
    
    # Return last values
    return theta_0, theta_1

print gradient_descent_2()
```

```python
a = np.arange(-5, 25, 0.1)
b = np.arange(-5, 35, 0.1)

x, y = np.meshgrid(a, b)
z = f(x, y)

x_plot = []
y_plot = []

for i in range(len(grad_x)):
    if i%100 == 0:
        x_plot.append(grad_x[i])
        y_plot.append(grad_y[i])

fig1, ax1 = plt.subplots(figsize=(20, 10))
ax1.contour(x, y, z)
ax1.scatter(0, 0, color='black', label="target")

for i in range(len(x_plot)):
    ax1.scatter(grad_x[i], grad_y[i], label="x=%.2f y=%.2f" % (grad_x[i], grad_y[i]))
    ax1.annotate("", xy=(grad_x[i+1], grad_y[i+1]), xytext=(grad_x[i], grad_y[i]), 
                 arrowprops=dict(arrowstyle="->", color='black', lw=.5))

ax1.set_title("Converging of the Gradient")
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend(loc='best')
```

```python
# Part 3

# Create random noise
n = 100
x = np.arange(n)
theta_0 = -3
theta_1 = .8
noise = np.random.normal(size=n)

# Define our function
def lm_f(theta_0_true, theta_1_true):
    return theta_0_true + theta_1_true*x + noise

y = lm_f(theta_0, theta_1)

def plot_linear_data(x, y):
    plt.figure(figsize=(12, 7))
    plt.title("Random Linear Data")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.scatter(x, y);
    plt.legend(['Hypothesis', 'Data'], loc='best')

plot_linear_data(x, y)
```
